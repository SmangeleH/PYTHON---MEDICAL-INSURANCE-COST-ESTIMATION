# Medical Insurance Cost Estimator

## Project Overview
This project explores how a simple formula can be used to estimate annual medical insurance costs based on personal factors such as age, sex, BMI, number of children, and smoking status.

It was originally completed as a Python syntax exercise and has been standardised into a cleaner format for GitHub presentation.


## Objectives
- practice Python syntax and arithmetic operations
- calculate a base insurance estimate
- test how small changes in personal factors affect the estimate
- present a beginner Python project in a professional format


```python
# Initial profile
age = 28
sex = 0          # 0 = female, 1 = male
bmi = 26.2
num_of_children = 3
smoker = 0       # 0 = non-smoker, 1 = smoker
```

## Insurance Cost Formula

The estimate is based on the following formula:

\[
insurance\_cost =
250 \times age
- 128 \times sex
+ 370 \times bmi
+ 425 \times num\_of\_children
+ 24000 \times smoker
- 12500
\]


```python
def estimate_insurance_cost(age, sex, bmi, num_of_children, smoker):
    return (
        250 * age
        - 128 * sex
        + 370 * bmi
        + 425 * num_of_children
        + 24000 * smoker
        - 12500
    )
```

```python
insurance_cost = estimate_insurance_cost(age, sex, bmi, num_of_children, smoker)
print(f"Base insurance cost: ${insurance_cost}")
```

## Scenario 1: Increase Age by 4 Years

```python
new_age = age + 4
new_insurance_cost = estimate_insurance_cost(new_age, sex, bmi, num_of_children, smoker)
change_in_insurance_cost = new_insurance_cost - insurance_cost

print(f"Change after increasing age by 4 years: ${change_in_insurance_cost}")
```

## Scenario 2: Increase BMI by 3.1

```python
new_bmi = bmi + 3.1
new_insurance_cost = estimate_insurance_cost(age, sex, new_bmi, num_of_children, smoker)
change_in_insurance_cost = new_insurance_cost - insurance_cost

print(f"Change after increasing BMI by 3.1: ${change_in_insurance_cost}")
```

## Scenario 3: Compare Female vs Male

```python
male_insurance_cost = estimate_insurance_cost(age, 1, bmi, num_of_children, smoker)
change_in_insurance_cost = male_insurance_cost - insurance_cost

print(f"Change for being male instead of female: ${change_in_insurance_cost}")
```
